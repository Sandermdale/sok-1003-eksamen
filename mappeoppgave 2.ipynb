En bedrift produserer biler. Produktfunksjonen til bedriften defineres slik  $f(L, a, R) = 2RL^{a}$, hvor:
- `L` er arbeidskraft, 
- `a` er produktiviteten til arbeiderne og 
- `R` er antall robotmaskiner

a) Lag en formel for produktfunksjonen til bedriften og plot den grafisk med ulike verdier av `L` på x-aksen. Anta `a=0.6` og `R=2`

```python
#definerer produktfunskjonen til bedriften
def product(L,a,R):
    return(2*R*L**a)
```

```python
#regner ut produktet ved 3000 ansatte og lagrer svaret som value
value=product(3000,0.6,2)
print(value)
```

```python
#plotter figuren til funksjonen 
x = np.linspace(0, 5000, 1000)#definerer x aksen 
plt.figure(figsize =(10, 6))
plt.plot(x, product(x,0.6, 2), color='blue')
plt.suptitle('Produktfunksjon ved 6 roboter',fontsize=24)
#plotter grafen med produkt funksjonen. 

plt.scatter(3000,value, color='red',
            label='Eksempel på antall biler ved 3000 ansatte')
#bruker resultatet fra ovenfor og plotter punktet i grafen 

plt.xlabel("Antall arbeidere",fontsize=18) 

plt.ylabel("Antall produserte biler",fontsize=18) 

plt.legend(loc='lower right',frameon=False) 

plt.annotate((3000,487.9), xy=(2000, 550),fontsize=10, color="red")

plt.show()#gir navn på akser og gir tittel til figuren 
```

b) anta at profittfunksjonen til denne bedriften er ${profit = f(L, a, R)p-wL-cR-K}$, hvor 
- `w` er månedslønnen til arbeiderne, 
- `c` er kostnaden for robotmaskinene 
- `K` er faste kostnader
- `p` er utsalgsprisen på bilene. 

Anta `a=0.6`,`R=6`, `p=300 000`, `w=100 000`, `c= 1 000 000` og `K=90 000 000`. Plot profittfunksjonen figurativt for antall arbeidere (`L`) mellom 0 og 10 000. Vis profitten i millioner (dvs at du må dele på 1 000 000)

```python
#definerer proffittfunksjone til bedriften 
def profit(L,a,R,p,w,c,K): 
    return (product(L,a,R)*p-w*L-c*R-K)

profit(2000,0.6,6,300000,100000,1000000,90000000) 
```

```python
#plotter grafen til profittfunksjonen
L = np.linspace(0,8000,100) 

fig,ax=plt.subplots(figsize=(10,8)) 

ax.set_ylabel('Fortjeneste i millioner',fontsize=18) 

ax.set_xlabel('Antall ansatte',fontsize=18) 

plt.plot(L,profit(L,0.6,6,300000,100000,1000000,90000000)/1000000,
         label='Fortjeneste med 6 maskiner', color="darkblue") 
fig.suptitle("Bedriftens profittfunkjson ved 6 robotmaskiner",fontsize=24)

ax.legend(loc='upper right',fontsize=12) 
```

d) finn profittmaksimum og optimal antall arbeidere ved hjelp av derivasjon med samme forutsetninger som i (1b). Bruk `sympy`-pakken til dette 

```python
L,a,R,p,w,c,K=sp.symbols("L,a,R,p,w,c,K") 
#gjør variablene til symboler, slik at man kan regne med dem
```

```python
#finne utt profittmaksimum 
profit(L,a,R,p,w,c,K) 
```

```python
#lager en ny variabel som heter B_profitt 
B_profit=sp.diff(profit(L,a,R,p,w,c,K),L)#dervierer profitt funksjonen med hensyn på L 

B_profit #sjekker at alt er rett 
```

```python
foc=sp.Eq(B_profit,0) #setter deretter likningen av profitt til =0 

foc #sjekker at alt stemmer 
```

```python
L_max=solve(foc,L)[0] #deretter løser vi denne likningen ved hjelp av solve funksjonen 

L_max #sjekker at alt stemmer 
```

```python
profit_max=profit(L_max,a,R,p,w,c,K) 
#setter deretter inn løsningen av å finne L(arbeidere) in i profitt funksjonen 
```

```python
num_dict={a:0.6,R:6,p:300000,w:100000,c:1000000,K:90000000} 
#lager en liste med alle variablene som er definert for oss. 
```

```python
L_max.subs(num_dict) 
#deretter løser vi likningen ved hjelp av listen vår og får dette svaret:
```

```python
num_dict[L]=L_max.subs(num_dict)#setter deretter inn svaret vi fikk over i listen vår og får en full liste 

num_dict#sjekker at alt stemmer 
```

```python
profit_max=float(profit(L,a,R,p,w,c,K).subs(num_dict)) 
#så regner vi ut den totale profitten med listen vår og profitt formelen 

profit_max #sjekker at alt er riktig. 
```
