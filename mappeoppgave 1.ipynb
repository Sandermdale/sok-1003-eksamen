c) Vis hva det samme lånet koster som annuitetslån, dvs differansen mellom alle terminbeløp og lånebeløp.<br> <br> Annuitetslån gir like terminbeløp hver måned, men renten utgjør en større del av dette beløpet i starten. Terminbeløpet for et annuitetslån er definert ved formelen:
$T = x*\frac{r/f} {(1-(1+(r/f))^{-t})}$, hvor x=lånebeløp, r = årlig rente, t = terminer, f= antall forrentinger





```python
lånebeløp=1000000
#definerer en funksjon for terminbeløpene for et annuitetslån
def terminbeløp_(x,t,r,f):
    return x*(r/f)/(1-(1+(r/f))**(-t)) 
annuitetslån=terminbeløp_(1000000,120,0.03,12)
annuitetslån_sum=annuitetslån*120
print(annuitetslån_sum-lånebeløp) #finner ut hva lånet koster som annuitetslån
```

Dvs, hvis t=15, har det gått 1 år og 3 mnd med terminer. Se forøvrig relevante formler i oppgave (a) 

<br> 

<b>Hint1:</b> Siden avdragene er like for alle måneder, kan det være lurt å definere det månedlige avdraget som en liste og gange det med antall perioder. </br>
<b>Hint2:</b> Siden vi er ute etter både rentekostnader og avdrag hver for seg, kan det være lurt å definere en funksjon for hver av dem. </br>


```python
r=0.03 
f=12 
x=1000000
t=np.arange(1,241) 
a=1000000/240
p=240

#definerer funksjon for avdragene
def s(x,p):
    return x/p
avdrag=s(x,p) 

#definerer funksjon for terminbeløp av serielån
def terminbeløp (x,a,t,r,f):
    return (x-a*(t-1))*r/f  
rentekostnader=4000+ terminbeløp(x,a,t,r,f) #definerer rentekostnadene

#bruker stackplot til å lage figuren
plt.stackplot(t,rentekostnader,colors='red', labels="Rentekostnader")
plt.stackplot(t,avdrag, colors='blue', labels="Avdrag")

#definerer egenskapene til figuren
plt.xlabel('Antall terminer')
plt.ylabel('Kr')
plt.legend()
plt.title("Utvikling i rentekostnader og avdrag over tid")
plt.show()
```
